```python
import joblib
import numpy as np
```

```python
model_path = r"C:\Users\Public\Downloads\5th Sem Project\MLModel\model.pkl"
scaler_path = r"C:\Users\Public\Downloads\5th Sem Project\MLModel\scaler.pkl"
```

```python

model = joblib.load(model_path)
scaler = joblib.load(scaler_path)
```

```python
# user_data = [Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age]
user_data = np.array([[4,110,92,0,0,37.6,0.191,30]])

user_data_scaled = scaler.transform(user_data)
```

```python
prediction = model.predict(user_data_scaled)

if prediction[0] == 1:
    print("\nResult: The patient has diabetes.")
else:
    print("\nResult: The patient does not have diabetes.")
```
